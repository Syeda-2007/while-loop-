```python
num = int(input("Enter a number: "))
i = 1
while i <= 10:
    print(num, "x", i, "=", num * i)
    i += 1
```

```python
N = int(input("Enter a positive integer: "))
sum = 0
i = 1
while i <= N:
    sum += i
    i += 1

print("The sum from 1 to", N, "is:", sum)
```

```python
num = int(input("Enter a number: "))
while num <= 0:
    print("Invalid! Please enter a positive number.")
    num = int(input("Enter a number: "))

print("You entered a positive number:", num)
```

```python
num = int(input("Enter a number: "))
num = abs(num)
count = 0
if num == 0:
    count = 1
else:
    while num > 0:
        num //= 10   
        count += 1
        
print("Number of digits:", count)
```

```python
total = 0
num = int(input("Enter a number (0 to stop): "))
while num != 0:
    total += num
    num = int(input("Enter a number (0 to stop): "))

print("Final total:", total)
```

```python
num = 101 
while True:
    if num % 7 == 0:
        print("First multiple of 7 greater than 100 is:", num)
        break
    num += 1
```

```python
secret_number = 25
attempts = 0
guess = int(input("Guess the number: "))
while guess != secret_number:
    attempts += 1
    if guess > secret_number:
        print("Too high!")
    else:
        print("Too low!")
    
    guess = int(input("Guess again: "))

attempts += 1  # count the correct attempt
print("Correct! You guessed it in", attempts, "attempt(s).")
```
